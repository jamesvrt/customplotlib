```python
from customplotlib.base import Customplotlib
```

```python
plt = Customplotlib(supp_colors=True)
```

```python
plt.plot(x=[1.0,2.0,3.0], y=[1.0,1.5,3.0], label='That')
plt.plot(x=[1.0,2.0,3.0], y=[3.0,2.0,1.0], label='This', title='Example Plot')
```
